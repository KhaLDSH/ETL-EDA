```python
import pandas as pd

def enforce_schema(df: pd.DataFrame) -> pd.DataFrame: return df.assign( order_ip=df["order_id"].astype("string"), amountl=pd.to_numeric(df["amount"], errors="coerce").astype("Float64") )

def missingness_report(df):
    return df.isna().sum().to_frame("n_missing").assign(
        p_missing=lambda t: t["n_missing"] / len(df)
    )

def add_missing_flags(df, cols):
    out = df.copy()
    for c in cols:
        out[f"{c}__isna"] = out[c].isna()
    return out

df = pd.DataFrame({
    "order_id": [1, 2, 3],
    "amount": ["10.5", "abc", 20]
})

dff = enforce_schema(df)

print(df)

dff = add_missing_flags(dff, ["amount"])
print(dff)
```
